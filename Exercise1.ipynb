A single linear equation with three inputs (i.e. linear_function_3D()) associates a value y with each point in a 3D space (x1
,x2
,x3
). Is it possible to visualize this? What value is at position (0,0,0)

```python
def linear_function_3D(x1, x2, x3):
    a = 1
    b = 2
    c = 3
    d = 4
    y = a*x1 + b*x2 + c*x3 + d
    return y
```

In this case, the value at position (0,0,0) is 4.

To visualize this, we will use 3D plotting library matplotlib in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
x1, x2 = np.meshgrid(x1, x2)
x3 = linear_function_3D(x1, x2, 0)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, x2, x3)

plt.show()
```

Write code to compute three linear equations with two inputs ( x1 ,  x2 ) using both the individual equations and the matrix form (you can make up any values for the inputs  i  and the slopes  ij .

We will define three linear equations with two inputs (x1, x2) as follows:

y1 = m11x1 + m12x2 + b1
y2 = m21x1 + m22x2 + b2
y3 = m31x1 + m32x2 + b3
We will compute compute these equations individually and in matrix form using Python and the numpy library:


```python
import numpy as np

# Define the slopes (m) and intercepts (b)
m11, m12, b1 = 1, 2, 3
m21, m22, b2 = 4, 5, 6
m31, m32, b3 = 7, 8, 9

# Define the inputs (x)
x1, x2 = 10, 11

# Compute the equations individually
y1 = m11*x1 + m12*x2 + b1
y2 = m21*x1 + m22*x2 + b2
y3 = m31*x1 + m32*x2 + b3

print(f"Individual computations: y1={y1}, y2={y2}, y3={y3}")

# Compute the equations in matrix form
M = np.array([[m11, m12], [m21, m22], [m31, m32]])  # slopes
x = np.array([x1, x2])  # inputs
b = np.array([b1, b2, b3])  # intercepts

y = np.dot(M, x) + b

print(f"Matrix computation: y={y}")
```

A function is convex if we can draw a straight line between any two points on the function, and this line always lies above the function. Similarly, a function is concave if a straight line between any two points always lies below the function. Is the exponential function convex or concave or neither?

Answer:

The exponential function, exp(x), is convex. This means that for any two points on the function, the line segment connecting them lies above the function. This property holds because the second derivative of the exponential function is always positive, which is a mathematical condition for convexity.

In mathematical terms, if f(x) is the exponential function, then the second derivative f''(x) = exp(x), which is always greater than 0 for all real x. Therefore, the exponential function is convex
